# üìò Listy w Pythonie

Listy to **podstawowy typ danych** w Pythonie s≈Çu≈ºƒÖcy do przechowywania wielu element√≥w w jednej strukturze.

Mo≈ºna w nich przechowywaƒá liczby, napisy, a nawet inne listy. SƒÖ **modyfikowalne (mutable)** i zachowujƒÖ kolejno≈õƒá element√≥w.

## 1Ô∏è‚É£ Tworzenie list
Listƒô tworzymy u≈ºywajƒÖc nawias√≥w kwadratowych `[]` lub funkcji `list()`.

```python
# Tworzenie list
liczby = [1, 2, 3, 4, 5]
owoce = ["jab≈Çko", "banan", "czere≈õnia"]
mieszana = [1, "tekst", True, 3.14]

print(liczby)
print(owoce)
print(mieszana)
```

## 2Ô∏è‚É£ Dostƒôp do element√≥w listy
- Indeksy zaczynajƒÖ siƒô od `0`.
- Mo≈ºemy te≈º u≈ºywaƒá indeks√≥w ujemnych (od ko≈Ñca listy).

```python
owoce = ["jab≈Çko", "banan", "czere≈õnia"]
print(owoce[0])   # pierwszy element
print(owoce[-1])  # ostatni element
```

## 3Ô∏è‚É£ Modyfikowanie listy
Mo≈ºemy zmieniaƒá elementy listy, dodawaƒá i usuwaƒá pozycje.

```python
owoce = ["jab≈Çko", "banan", "czere≈õnia"]
owoce[1] = "gruszka"  # zmiana elementu
owoce.append("≈õliwka")  # dodanie na koniec
owoce.remove("jab≈Çko")  # usuniƒôcie elementu
print(owoce)
```

## 4Ô∏è‚É£ Najwa≈ºniejsze metody list
- `append(x)` ‚Äì dodaje element na koniec
- `insert(i, x)` ‚Äì wstawia element na wskazanej pozycji
- `remove(x)` ‚Äì usuwa pierwsze wystƒÖpienie elementu
- `pop(i)` ‚Äì usuwa element o danym indeksie i go zwraca
- `sort()` ‚Äì sortuje listƒô
- `reverse()` ‚Äì odwraca kolejno≈õƒá element√≥w
- `len(lista)` ‚Äì zwraca d≈Çugo≈õƒá listy

```python
liczby = [4, 1, 7, 3]
liczby.append(9)
liczby.sort()
print("Posortowane:", liczby)
liczby.reverse()
print("Odwr√≥cone:", liczby)
print("D≈Çugo≈õƒá listy:", len(liczby))
```

## 5Ô∏è‚É£ Iteracja po li≈õcie
Mo≈ºemy przechodziƒá po elementach listy pƒôtlƒÖ `for`.

```python
owoce = ["jab≈Çko", "gruszka", "≈õliwka"]
for owoc in owoce:
    print(f"Owoc: {owoc}")
```

## 6Ô∏è‚É£ Listy zagnie≈ºd≈ºone
Listy mogƒÖ zawieraƒá inne listy ‚Äì to tzw. **listy zagnie≈ºd≈ºone**.

```python
macierz = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
print(macierz[0])      # pierwszy wiersz
print(macierz[1][2])   # element z drugiego wiersza, trzeciej kolumny
```
